```python
class Pokemon:
    def __init__(self, name, level, health, max_health, tipe, is_knocked_out):
        self.name = name
        self.level = level
        self.health = health
        self.max_healt = max_health
        self.tipe = tipe
        self.is_knocked_out = is_knocked_out
    
    def lose_health(self, damage):
        self.health -= damage
        if self.health > 0:
            print('Now {} have {} of HP'.format(self.name, self.health))
        else:
            print('{} has fainted'.format(self.name))
            self.fainted()
        
    def gain_health(self, regain):
        if self.is_knocked_out == False:
            self.health += regain
            if self.health > self.max_health:
                print('Your pokemon has restored his full health')
                self.health = self.max_health
            else:
                print('Now {} have {} of HP'.format(self.name, self.health))
        else:
            print('This does not work on fainted pokemons')
     
    def fainted(self):
        self.is_knocked_out = True
        print('{} can no longer fight'.format(self.name))
    
    def attack(self, Pokemon):
        switch ={
        'water fire': 2.0,
        'fire water': 0.5,
        'fire grass': 2.0,
        'grass fire': 0.5,
        'water grass': 0.5,
        'grass water': 2.0,
        'electricity water': 2.0,
        'electricity grass': 1.0,
        'electricity fire': 1.0,
        'water electricity': 0.5,
        'grass electricity': 1.0,
        'fire electricity': 1.0,
        'fire fire': 1.0,
        'water water':1.0,
        'grass grass':1.0,
        'electricity electricity': 1.0}
        attack_value = 3
        modiffier = str(self.tipe) + ' ' + str(Pokemon.tipe)
        attack_modiffier = switch[modiffier]
        damage = attack_value * attack_modiffier
        Pokemon.lose_health(damage)
        print('{} gets hit by {} given damage {}'.format(Pokemon.name, self.name, damage))

class Trainer:
    def __init__(self, name, n_pokemons, max_pokemons, potions, active_pokemons, current_pokemon): 
        self.name = name
        self.n_pokemons = n_pokemons
        self.max_pokemons = max_pokemons
        self.potions = potions
        self.active_pokemons = active_pokemons
        self.current_pokemon = current_pokemon
    
    def attack_trainer(self, Trainer):
        trainer_pokemon = Trainer.current_pokemon
        self.current_pokemon.attack(trainer_pokemon)
    
    def cure_pokemon(self, Pokemon):
        regain = 20
        self.potions -= 1
        Pokemon.gain_health(regain)
    
    def options(self):
        switch = {
            1 : self.attack_trainer(),
            2 : self.cure_pokemon()}
        response = input('1 attack , 2 cure')
        switch[response]
        
charmander = Pokemon('Charmander', 5, 20, 20, 'fire', False)
bulbasaur = Pokemon('Bulbasaur', 5, 20, 20, 'grass', False)
squirtle = Pokemon('Squirtle', 5, 20, 20, 'water', False)
pikachu = Pokemon('Pikachu', 5, 20, 20, 'electricity', False)

red = Trainer('Red', 2, 6, 3, [pikachu, charmander], pikachu)
blue = Trainer('Blue', 2, 6, 3, [squirtle, bulbasaur], squirtle)

red.attack_trainer(blue)
```
